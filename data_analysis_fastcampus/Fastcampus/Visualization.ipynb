```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,10)
```

```python
y = x **2
```

```python
plt.plot(x,y, 'g')
plt.title('Plot Title')
plt.xlabel('X_label')
plt.ylabel('Y_label')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.7, 0.7])

plt.plot(x,y, 'g')
plt.title('Plot Title')
plt.xlabel('X_label')
plt.ylabel('Y_label')
```

```python
plt.plot(y,x, 'b')
plt.title('Plot Title')
plt.xlabel('X_label')
plt.ylabel('Y_label')
```

```python
fig = plt.figure()
axes1= fig.add_axes([0.1,0.1,0.9,0.9])
axes2= fig.add_axes([0.3,0.3,0.3,0.3])

axes1.plot(x,y, 'g')
plt.title('Plot Title')
plt.xlabel('X_label')
plt.ylabel('Y_label')

axes2.plot(y,x, 'b')
plt.title('Plot Title')
plt.xlabel('X_label')
plt.ylabel('Y_label')
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'g')
plt.title('Plot Title')
plt.xlabel('X_label')
plt.ylabel('Y_label')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'r')
axes[0].set_xlabel('x')
axes[1].plot(y,x,'b')
```

```python
fig.savefig('my_name.jpg')
```
