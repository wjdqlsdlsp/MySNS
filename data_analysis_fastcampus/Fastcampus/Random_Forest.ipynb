```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mem = pd.read_csv("member.csv")
tran = pd.read_csv("transaction.csv")
```

```python
mem
```

```python
tran
```

```python
data = tran.head(1000)
```

```python
data.groupby('id').sum().reset_index()
```

```python
tran['avg_price'] = tran['total_amount'] / tran['num_item']
```

```python
tran_mean = tran.groupby('id').mean()
```

```python
tran_cnt = tran['id'].value_counts()
```

```python
tran_mean
```

```python
tran_cnt
```

```python
tran_df =pd.concat([tran_mean, tran_cnt],axis=1)
```

```python
tran_df = tran_df.rename(columns={'id':'count'})
```

```python
tran_df
```

```python
mem.set_index('id', inplace=True)
```

```python
mem
```

```python
data = mem.join(tran_df)
```

```python
data.isna().sum()
```

```python
data
```

```python
data = pd.get_dummies(data, columns=['zip_code', 'channel'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('conversion',axis=1),data['conversion'] ,test_size=.3, random_state=100)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(max_depth = 10,random_state=100)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(max_depth=10, random_state=100)
```

```python
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_test)
```

```python
df_pred = pd.DataFrame(pred)
```

```python
df_pred.columns = ["value"]
```

```python
def conv(value):
    if value >= 0.5:
        return 1
    else:
        return 0
```

```python
df_pred["conv"] = df_pred['value'].apply(conv)
```

```python
df_pred
```

```python
accuracy_score(df_pred["conv"],y_test)
```

```python
confusion_matrix(y_test, df_pred["conv"])
```
