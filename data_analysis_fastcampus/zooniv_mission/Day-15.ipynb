```python
# 필요한 Module을 Import 하세요
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 주어진 데이터를 불러오세요
df = pd.read_csv("Day15_heart.csv")
```

### 1. age, thalach 컬럼을 scatter plot으로 표현해보세요 (sex 컬럼으로 색상을 나타내세요)


```python
# 정답 Code
sns.scatterplot(data= df, x='age',y='thalach',hue='sex')
```

```python
# 아래와 같은 형태가 나와야합니다.
```

### 2. age 컬럼을 distribution plot을 그리되, 분포곡선(kde)는 나타나지 않게 하세요.

```python
# 정답 Code
sns.distplot(df['age'],kde=False)
```
