```python
# 필요한 Module을 Import 하세요
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 주어진 데이터를 불러오세요
data = pd.read_csv("Day12_Time.csv",index_col =0)
```

### 1. AR 모델을 이용하여 타임시리즈 예측 모델을 만드세요. (이전 12개월 데이터를 학습하도록 파라미터 설정)

```python
# 정답 Code
# Hint: date는 index로 들어가있어야합니다.
from statsmodels.tsa.ar_model import AutoReg
```

```python
model = AutoReg(data, lags = 12)
model_fit = model.fit()
```

### 2. 학습된 AR모델을 이용하여, 향후 30개월 데이터를 예측하고 결과물을 Line plot으로 그려보세요

```python
# 정답 Code
pred = model_fit.predict(start = len(data), end = len(data)+29)
```

```python
pred.plot(figsize=(15,10))
```
