```python
# 필요한 Module을 Import 하세요
import pandas as pd
import string
import matplotlib.pyplot as plt
```

```python
# 주어진 데이터를 불러오세요
data = pd.read_csv("Day13_Vaccination.csv")
```

### 1. text 컬럼에서 불필요한 기호를 제거하세요. (다음 작업을 위해 결과물을 기존 데이터에 덮어쓰세요)

```python
# 정답 Code
def delete_text(text):
    result = []
    for i in text:
        if i not in  string.punctuation:
            result.append(i)
    result = ''.join(result)
    return result
```

```python
data['text'] = data['text'].apply(delete_text)
```

### 2. text 컬럼에서 상투적인 표현을 제거하세요. (다음 작업을 위해 결과물을 기존 데이터에 덮어쓰세요)

```python
# 정답 Code
from nltk.corpus import stopwords
```

```python
def delete_text2(text):
    result = []
    for i in text.split():
        if i.lower() not in  stopwords.words('english'):
            result.append(i.lower())
    return result

data['text'] = data['text'].apply(delete_text2)
```

```python
data['text']
```

### 3. 정리된 텍스트 데이터를 가지고 워드클라우드를 만들어보세요.

```python
# 정답 Code
word_split = []
for i in range(len(data)):
    for j in data.iloc[i]['text']:
        word_split.append(j)
```

```python
len(word_split)
```

```python
from nltk.probability import FreqDist

plt.figure(figsize=(20,10))
FreqDist(word_split).plot(30)
```
